```python
unemployment_oecd = pd.read_excel('unemployment_2023.xlsx', sheet_name='Sheet2')
```

```python
chart = alt.Chart(unemployment_oecd).mark_circle(opacity=0.8, stroke='black', strokeWidth=1, strokeOpacity=0.4).encode(
    y=alt.Y('country:N').title('Country'),
    x=alt.X('unemployment:Q').title('Unemployment Rate (%)'),
    size=alt.Size('population:Q', scale=alt.Scale(range=[0, 2500]), title='Population'),
    color=alt.Color('country:N', legend=None),
    tooltip=['country', 'region', 'unemployment', 'population']
).properties(
    width=800,
    height=400,
    title=alt.Title(
        text="Unemployment Rates and Population by Region",
        anchor='start'
    )
).configure_axis(
    grid=False
).configure_view(
    stroke=None
)

chart
```
